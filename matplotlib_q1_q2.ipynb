```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
fin = pd.read_csv('q1_q2_2019.csv')
```

```python
#Calculate the average ride_length for members and casual riders. Try rows = member_casual; Values = Average of ride_length.

y = np.array(fin.groupby(['usertype'])['minutes'].mean())
```

```python
x = np.array(["Customer", "Subscriber"])
plt.bar(x, y)
plt.ylabel("minutes") 
plt.show()
```

```python

y2y= fin.groupby(['usertype'])['minutes'].mean()
fin = fin.replace(np.nan, 0)
plt.ylabel("minutes")
y2y.plot(kind='bar', figsize=(10, 7))
```

```python
#Calculate the average ride_length for users by day_of_week. Try columns = day_of_week; Rows = member_casual; Values = Average of ride_length.
lo=fin.groupby(['usertype','day_of_week'])['minutes'].mean()

fin = fin.replace(np.nan, 0)

lo.plot(kind='bar', figsize=(10, 7))
```

```python
#Calculate the average ride_length for users by month

k = fin.groupby(['usertype','month'])['minutes'].mean()

fin = fin.replace(np.nan, 0)

k.plot(kind='bar', figsize=(10, 7))
```
